## Tokenizar Texto

```python
# LEER TEXTO DESDE URL
from bs4 import BeautifulSoup
import urllib.request
```

```python
response = urllib.request.urlopen('https://es.wikipedia.org/wiki/Finanzas') #Obtiene el contenido HTML de la pgina web http://librefinanciero.com/
 
html = response.read()
 
soup = BeautifulSoup(html,"html5lib") # Limpia el texto HTML capturado
 
text = soup.get_text(strip=True)
 
print (text)
```

```python
# Crear Tokes por palabras
from nltk.tokenize import word_tokenize
```

```python
tokens = word_tokenize(text, "spanish")
```

```python
tokens=[word.lower() for word in tokens if word.isalpha()] #Eliminar los signos de puntuacin
print(tokens)
```

```python
#Verificar frecuencia de palabras 
freq = nltk.FreqDist(tokens)
for key,val in freq.items():
    print(str(key)+ ':'+ str(val))
```

```python
# Visualizar los tokens
import matplotlib.pyplot as plot
import seaborn as sns
```

```python
sns.set()
freq.plot(30,cumulative=False)
```

## 2. Eliminar Palabras de Parada

```python
from nltk.corpus import stopwords
```

```python
clean_tokens = tokens[:]
for token in tokens:
    if token in stopwords.words('spanish'):
        clean_tokens.remove(token)
```

```python
print(clean_tokens)
```

```python
#Verificar frecuencia de palabras 
freq_clean = nltk.FreqDist(clean_tokens)
for key,val in freq_clean.items():
    print(str(key)+ ':'+ str(val))
```

```python
sns.set()
freq_clean.plot(30,cumulative=False)
```

### 3a. Obtener Sinnimos

```python
from nltk.corpus import wordnet
```

```python
synonyms = []
for syn in wordnet.synsets('investment'):
    for lemma in syn.lemmas():
        synonyms.append(lemma.name())
print(synonyms)
```

```python
#Reemplazar tokens sinnimos
for ind,sin in enumerate(synonyms):
    clean_tokens_sin = [word.replace(synonyms[ind],'investment') for word in clean_tokens]
```

```python
#Reemplazar tokens sinnimos manual
sinonimos = ['valor', 'value']
for ind,sin in enumerate(sinonimos):
    clean_tokens_sin = [word.replace(sinonimos[ind],'valor') for word in clean_tokens]
```

```python
#Recalcular frecuencia de palabras con sinnimos agregados
freq_clean_sin = nltk.FreqDist(clean_tokens_sin)
for key,val in freq_clean_sin.items():
    print (str(key) + ':' + str(val))
```

```python
#Visualizar Tokens
freq_clean_sin.plot(30, cumulative=False)
```

### 3b. Obtener Antnimos

```python
from nltk.corpus import wordnet
 
antonyms = []
 
for syn in wordnet.synsets("good"):
 
    for l in syn.lemmas():
 
        if l.antonyms():
 
            antonyms.append(l.antonyms()[0].name())
 
print(antonyms)
```
