## Tokenizar Texto

```python
# LEER TEXTO DESDE URL
from bs4 import BeautifulSoup
import urllib.request
```

```python
response = urllib.request.urlopen('https://es.wikipedia.org/wiki/Finanzas') #Obtiene el contenido HTML de la página web http://librefinanciero.com/
 
html = response.read()
 
soup = BeautifulSoup(html,"html5lib") # Limpia el texto HTML capturado
 
text = soup.get_text(strip=True)
 
print (text)
```

```python
# Crear Tokes por palabras
from nltk.tokenize import word_tokenize
```

```python
tokens = word_tokenize(text, "spanish")
```

```python
tokens=[word.lower() for word in tokens if word.isalpha()] #Eliminar los signos de puntuación
print(tokens)
```

```python
#Verificar frecuencia de palabras 
freq = nltk.FreqDist(tokens)
for key,val in freq.items():
    print(str(key)+ ':'+ str(val))
```
